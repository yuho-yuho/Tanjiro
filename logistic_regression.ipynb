## Test Logistic Regression Method with Sigmoid Function -- Classification

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

data = pd.read_csv('./logr/Diabetes_Data.csv')
data["Gender"] = data["Gender"].map({"男生": 1, "女生": 0})

data
```

```python
from sklearn.model_selection import train_test_split

x = data[["Age", "Weight", "BloodSugar", "Gender"]]
y = data["Diabetes"]

# separate training & testing sets: 20% used for testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=86) 

x_train = x_train.to_numpy()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x_train) # only use training data, no testing data!!

x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)

x_train
```

```python
import numpy as np
def sigmoid(z):
    return 1/(1+np.exp(z))
```

```python
w = np.array([1, 2, 3, 4])
b = 1.0

z = (w*x_train).sum(axis=1) + b
sigmoid(z)
```
